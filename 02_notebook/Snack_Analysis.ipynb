```python
# Read Excel files using Pandas

menu_pd = pd.read_excel("../data/menu.xlsx")

starbucks_drink_expanded_pd = pd.read_excel("../data/starbucks_drinkMenu_expanded.xlsx")

starbucks_drinks_pd = pd.read_excel("../data/starbucks-menu-nutrition-drinks.xlsx")

starbucks_food_pd = pd.read_excel("../data/starbucks-menu-nutrition-food.xlsx")
```

```python
# Convert all columns to string
menu_pd = menu_pd.astype(str)
starbucks_drink_expanded_pd = starbucks_drink_expanded_pd.astype(str)
starbucks_drinks_pd = starbucks_drinks_pd.astype(str)
starbucks_food_pd = starbucks_food_pd.astype(str)

# Convert to Spark DataFrames
menu_df = spark.createDataFrame(menu_pd)

starbucks_drink_expanded_df = spark.createDataFrame(starbucks_drink_expanded_pd)

starbucks_drinks_df = spark.createDataFrame(starbucks_drinks_pd)

starbucks_food_df = spark.createDataFrame(starbucks_food_pd)
```

```python
from pyspark.sql.functions import col

menu_clean = (
    menu_df
    .withColumn("Calories", col("Calories").cast("int"))
    .withColumn("Total Fat", col("Total Fat").cast("double"))
    .withColumn("Sodium", col("Sodium").cast("int"))
    .withColumn("Carbohydrates", col("Carbohydrates").cast("int"))
    .withColumn("Sugars", col("Sugars").cast("int"))
    .withColumn("Protein", col("Protein").cast("int"))
)

menu_clean.printSchema()
```

```python
##Which McDonald's food categories have the highest average calories?
from pyspark.sql.functions import avg

menu_clean.groupBy("Category") \
    .agg(avg("Calories").alias("Average Calories")) \
    .orderBy("Average Calories", ascending=False) \
    .show(truncate=False)
```

```python
##let's find the Top 10 highest-calorie items
menu_clean.select(
    "Item",
    "Category",
    "Calories"
).orderBy("Calories", ascending=False).show(10, truncate=False)
```

```python
##find the healthiest (lowest-calorie) items
menu_clean.select(
    "Item",
    "Category",
    "Calories"
).orderBy("Calories").show(10, truncate=False)
```
